```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import numpy as np

data=pd.read_csv("C:\\Users\\lenovo\\OneDrive\\Desktop\\Datasets\\Salary_Data.csv")

print(data.isna().sum())


print(data.fillna(0))


x=data[['YearsExperience']]
y=data[['Salary']]
# min_max_scaler = MinMaxScaler().fit(x)
# x_norm = min_max_scaler.transform(x)
# x_norm
#Splitting data with model
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2,random_state=42)
model = LinearRegression()
model.fit(x_train, y_train)
#Predicating y with x_test
y_pred = model.predict(x_test)
new_data = np.array([9]).reshape(-1, 1)
new_data
a = model.predict(new_data)
print("predicated marks",a)
#using scatter plot ploting x_test,y_tset and predicated y_pred with blue and red color
plt.scatter(x,y, color = "blue")
plt.plot(x_test, y_pred, color='red', linewidth=2)
plt.ylabel('Salary')
plt.title('Experience Years VS Salary')
plt.show()


#using scatter plot for comparing
plt.scatter(x_test,y_test, color = "blue")
plt.scatter(x_test, y_pred, color='red')
plt.xlabel('Experience Years')
plt.ylabel('Salary')
plt.title('Experience Years VS Salary')
plt.show()

train=model.score(x_train,y_train)
test=model.score(x_test,y_pred)
print("train",train)
print("test",test)








```
